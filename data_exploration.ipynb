```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better looking plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Load the data
print("Loading data...")
df = pd.read_csv('train.csv')

# Define the specific columns you want to analyze
selected_columns = [
    "Component1_fraction", "Component2_fraction", "Component3_fraction",
    "Component4_fraction", "Component5_fraction", "Component2_Property1", 
    "Component5_Property1", "Component5_Property2", "Component3_Property3", 
    "Component5_Property3", "Component5_Property4", "Component5_Property6", 
    "Component2_Property8", "Component5_Property8", "Component2_Property9",
    "Component5_Property9", "Component2_Property10", "Component5_Property10"
]

# Filter to only include columns that exist in the dataset
available_columns = [col for col in selected_columns if col in df.columns]
missing_columns = [col for col in selected_columns if col not in df.columns]

print(f"Available columns: {len(available_columns)}")
print(f"Missing columns: {missing_columns}")

# Create histograms
fig, axes = plt.subplots(3, 6, figsize=(20, 12))
axes = axes.flatten()

for i, column in enumerate(available_columns):
    if i < len(axes):
        # Create histogram
        axes[i].hist(df[column], bins=30, alpha=0.7, edgecolor='black')
        axes[i].set_title(f'{column}', fontsize=10, fontweight='bold')
        axes[i].set_xlabel('Value')
        axes[i].set_ylabel('Frequency')
        axes[i].grid(True, alpha=0.3)
        
        # Add statistics text
        mean_val = df[column].mean()
        std_val = df[column].std()
        axes[i].text(0.02, 0.98, f'Mean: {mean_val:.3f}\nStd: {std_val:.3f}', 
                    transform=axes[i].transAxes, verticalalignment='top',
                    bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

# Hide unused subplots
for i in range(len(available_columns), len(axes)):
    axes[i].set_visible(False)

plt.tight_layout()
plt.suptitle('Histograms of Selected Features', fontsize=16, fontweight='bold', y=1.02)
plt.show()

# Print summary statistics
print("\n" + "="*60)
print("SUMMARY STATISTICS")
print("="*60)

for column in available_columns:
    print(f"\n{column}:")
    print(f"  Mean: {df[column].mean():.4f}")
    print(f"  Std:  {df[column].std():.4f}")
    print(f"  Min:  {df[column].min():.4f}")
    print(f"  Max:  {df[column].max():.4f}")
    print(f"  Range: {df[column].max() - df[column].min():.4f}")

# Save the plot
plt.savefig('selected_features_histograms.png', dpi=300, bbox_inches='tight')
print(f"\nHistogram saved as 'selected_features_histograms.png'")
```
